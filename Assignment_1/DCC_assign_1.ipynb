```python
import numpy as np
import pandas as pd
a=pd.read_csv("/content/drive/MyDrive/dcc/dcc - Sheet1.csv")
```

1. What is the average score of females in the class across all subjects?



```python
ans1=a[a['Gender']=='Female'][['Mathematics','Science','English']].mean().mean()
print(ans1)
```

2. How many students scored above the median score in Mathematics?


```python
ans2=(a['Mathematics']>a['Mathematics'].median()).sum()
print(ans2)
```

3. What is the mode of scores achieved by students in Science?


```python
ans3=a['Science'].mode()[1]
print(ans3)
```

4. How many male students scored below the mean score in English?


```python
ans4=(a[a['Gender']=='Male']['English']<a['English'].mean()).sum()
print(ans4)
```

5. What is the total score obtained by students aged 15 in the class?


```python
ans5=a[a['Age'] == 15][['Mathematics', 'Science', 'English']].sum().sum()
print(ans5)
```

6. How many female students scored between 85 and 90 in any subject?


```python
ans6 = a.loc[
    ((a['Gender']=='Female')&(((a['Mathematics']>=85)&(a['Mathematics']<=90))|((a['Science']>=85)&(a['Science']<=90))|((a['English']>=85)&(a['English']<=90))))
]
print(len(ans6))
```

7. What is the average score of students whose names start with 'A'?


```python
ans7=a[a['Name'].str[0] == 'A'][['Mathematics', 'Science', 'English']].mean().mean()
print(ans7)
```

8. How many students scored exactly 88 in any subject?


```python
a1=((a['Mathematics'] == 88)).sum()
a2=((a['English'] == 88)).sum()
a3=((a['Science'] == 88)).sum()
a4=((a['Mathematics'] == 88) & (a['Science'] == 88)).sum()
a5=((a['Mathematics'] == 88) & (a['English'] == 88)).sum()
a6=((a['English'] == 88) & (a['Science'] == 88)).sum()
a7=((a['Mathematics'] == 88) & (a['English'] == 88)&(a['Science'] == 88)).sum()
ans8=a1+a2+a3-a4-a5-a6+a7
print(ans8)
```

9. What is the median score of male students in the class?


```python
ans9=a[a['Gender'] == 'Male'][['Mathematics', 'Science', 'English']].median().median()
print(ans9)
```

10. How many female students scored above the mean score in Mathematics?

```python
ans10=(a[a['Gender'] == 'Female']['Mathematics']>a[a['Gender']=='Female']['Mathematics'].mean()).sum()
print(ans10)
```
